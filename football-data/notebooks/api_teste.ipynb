```python
import requests
import pandas as pd

headers = {
    'X-Auth-Token': '8ec5d7843d1c4532a3a330acfabceb65'
}

url = "https://api.football-data.org/v4/competitions/PL/matches?season=2024"
response = requests.get(url, headers=headers)

if response.status_code == 200:
    data = response.json()
    
    partidas = []
    for match in data['matches']:
        jogo = {
            'match_id': match['id'],
            'data_jogo': match['utcDate'],
            'status': match['status'],
            'liga_id': match['competition']['id'],
            'liga_nome': match['competition']['name'],
            'time_mandante_id': match['homeTeam']['id'],
            'time_mandante_nome': match['homeTeam']['name'],
            'time_visitante_id': match['awayTeam']['id'],
            'time_visitante_nome': match['awayTeam']['name'],
            'placar_mandante': match['score']['fullTime']['home'],
            'placar_visitante': match['score']['fullTime']['away']
        }

        partidas.append(jogo)

    df = pd.DataFrame(partidas)
    print(df.head())

    # Salvar em Excel
    df.to_excel('premier_league_matches_2024.xlsx', index=False)
    print("Arquivo Excel salvo com sucesso!")
else:
    print(f"Erro na requisição: {response.status_code}")
```
